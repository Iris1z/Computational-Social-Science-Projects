**Explaination and Analysis of My Decision**

- **Consideration of the purpose of this proj - Prioritization**  
  The primary reason for my decision to retain all 3 categories is grounded in the purpose of this analysis: to help the Department of Public Health prioritize establishments for audits given limited resources. By distinguishing between these 3 categories, the department gains the flexibility to make informed decisions based on resource availability.
  - When resources are highly constrained, they can focus solely on establishments predicted to fail inspections (Fail).
  - When more resources are available, they can also audit establishments classified as Pass w/ Conditions, as these represent borderline cases that may warrant attention.  

    This approach enables more nuanced analysis of outcomes, allowing the model to reflect real-world complexities and provide actionable insights across different levels of compliance.

- **Potential Concerns with This Decision**  
  While retaining three categories offers flexibility and granularity, I acknowledge certain drawbacks:
	1.	Class Imbalance: Multi-class classification can introduce imbalances between classes (e.g., Pass may dominate), potentially leading to poor model performance.
	2.	Model Complexity: Predicting three outcomes is more complex than a binary classification, which might reduce the overall accuracy or reliability of the model.
	3.	Trade-offs in Simplicity and Performance: Binarizing the outcome (e.g., grouping Fail and Pass w/ Conditions together) could lead to a more balanced dataset and simplify the task, which might improve model performance.

```python
# Necessary imports
# -----------

from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from xgboost import XGBClassifier
from sklearn.model_selection import StratifiedKFold, GridSearchCV
from sklearn.metrics import classification_report, accuracy_score
```

```python
# Define Stratified K-Fold Cross-Validation
skf = StratifiedKFold(n_splits=5, shuffle=True, random_state=42)
```

```python
# Placeholder for results
model_results = {}
```

**K-Fold type selection explanation**

I chose Stratified K-Fold Cross-Validation instead of standard K-Fold Cross-Validation because of the likely class imbalance in the dataset.
- In food inspection datasets, classes like 'Pass' are often much more common than 'Fail' or 'Pass w/ Conditions'. Using standard K-Fold could lead to some folds having very few (or even none) of the minority classes, resulting in unreliable model evaluations.
- Stratified K-Fold ensures that each fold contains the same class proportions as the overall dataset, allowing the model to be trained and tested on data representative of the real-world distribution.

**Selection of Evaluation Metric for GridSearchCV (& Validation)**
$$  \text{Recall} = \frac{\text{True Positives}}{\text{True Positives + False Negatives}} $$
where positive stands for fail; negtive stands for negtive;  

When selecting an evaluation metric for GridSearchCV and also conducting predictive auditing, it's essential to consider the specific purpose of the task. In this case, the purpose of the analysis is to help prioritize audits, with an emphasis on identifying high-risk establishments (Fail and potentially Pass w/ Conditions). My selection goes with recall. Below are some key considerations:
- Severe consequences of false negatives
  - Missing high-risk establishments could lead to serious public food health risks, making it critical to minimize false negtives. In this sense, recall meatures the proportion of actual high-risk establishments correctly identified, alighing with the goal.
- Class imbalance
  - The dataset has more Pass than Fail or Pass w/ Conditions, accuracy might not be suitable, as it would overemphasize the majority class. On the contrary, recall is better suited to handle imbalance in dataset as it focuses on identifying positive cases.

### Model 1/ Logistic Regression

- **logic**: A linear model that predicts probabilities for each catagery and works well when features and outcomes have a linear relationship.
- **pros**: Simple, interpretable, and computationally efficient.
- **cons**: May have limited performance with complex or non-linear relationships.
- **selection reason**: This model can acts as a baseline to compare against more complex models. 

```python
label_encoder = LabelEncoder()
y_encoded = label_encoder.fit_transform(y)

#
# Linear regression 
#-----------

# specify the hyperparameters
lr_model = LogisticRegression(random_state=10, max_iter=1000)

lr_param_grid = {'C': np.logspace(0, 1, 10),
                 'solver': ['liblinear', 'lbfgs']}

# execute the grid search
lr_grid = GridSearchCV(lr_model,
                       lr_param_grid,
                       cv=skf, 
                       scoring='recall_weighted',
                       verbose=1)

# now fit the tuning on the training data
lr_grid.fit(X, y_encoded)

# print the best performing model
model_results['Logistic Regression'] = {'best_score': lr_grid.best_score_,
                                        'best_params': lr_grid.best_params_,
                                        'model': lr_grid.best_estimator_,}

print("Best Parameters (Logistic Regression):", lr_grid.best_params_)
print("Best Recall:", lr_grid.best_score_)
```

###  Model 2/ Random Forest

- **logic**: An ensemble of decision trees that averages their predictions to improve accuracy and reduce overfitting.
- **pros**: Handles non-linear relationships and works well with categorical and numerical data.
- **cons**: Can be computationally expensive with many trees.
- **selection reason**: This model excels at identifying important features and works well for tabular data like inspections. 

```python
#
# Random Forest 
#-----------

rf_model = RandomForestClassifier(random_state=10)
```

```python
rf_model.get_params()
```

```python
# specify the hyperparameters
rf_param_grid = {'n_estimators': [100, 200],
                 'max_depth': [None, 10, 20],
                 'min_samples_split': [2, 5]}

# execute the grid search
rf_grid = GridSearchCV(rf_model,
                       rf_param_grid,
                       cv=skf, 
                       scoring='recall_weighted',
                       verbose=1)

# now fit the tuning on the training data
rf_grid.fit(X, y_encoded)

# print the best performing model
model_results['Random Forest'] = {'best_score': rf_grid.best_score_,
                                  'best_params': rf_grid.best_params_,
                                  'model': rf_grid.best_estimator_,}

print("Best Parameters (Random Forest):", rf_grid.best_params_)
print("Best Recall:", rf_grid.best_score_)
```

### Model 3/ XGBoost

- **logic**: Gradient boosting algorithm that optimizes for performance by building trees sequentially to minimize errors.
- **pros**: Highly accurate and efficient for tabular data.
- **cons**: Requires careful hyperparameter tuning and is less interpretable than simpler models.
- **selection reason**: This model has excellent performance for classification tasks, especially when prioritizing important inspections. 

```python
#
# XGBoost
#-----------

xgb_model = XGBClassifier(random_state=10, use_label_encoder=False, eval_metric='logloss')

# Processing data
class_mapping = {'Pass': 0, 'Fail': 1, 'Pass w/ Conditions': 2}
y_encoded = y.map(class_mapping)
```

```python
xgb_model.get_params()
```

```python
# specify the hyperparameters
xgb_param_grid = {'n_estimators': [100, 200, 400], # 300
                 'learning_rate': [0.007, 0.1, 0.2], # 0.01
                 'max_depth': [3, 6, 10]}

# execute the grid search
xgb_grid = GridSearchCV(xgb_model,
                        xgb_param_grid,
                        cv=skf,
                        scoring='recall_weighted',
                        verbose=1)

# now fit the tuning on the training data
xgb_grid.fit(X, y_encoded)

# print the best performing model
model_results['XGBoost'] = {'best_score': xgb_grid.best_score_,
                            'best_params': xgb_grid.best_params_,
                            'model': xgb_grid.best_estimator_,}

print("Best Parameters (XGBoost):", xgb_grid.best_params_)
print("Best Recall:", xgb_grid.best_score_)
```

```python
from sklearn.model_selection import cross_validate, cross_val_score
from sklearn.metrics import accuracy_score, recall_score, precision_score, f1_score, make_scorer
```

```python
print("\nValidation Metrics for All Models:")

# define the scoring metrics
scoring = {'accuracy': make_scorer(accuracy_score),
           'recall': make_scorer(recall_score, average='weighted'),
           'precision': make_scorer(precision_score, average='weighted'),
           'f1': make_scorer(f1_score, average='weighted')}

# For loop to loop thru each model
for model_name, model_data in model_results.items():
    print(f"\nMetrics for {model_name}:")
    best_model = model_data['model']
    
    # perform cross-validation with 5-fold and return the trained estimators
    cv_results = cross_validate(best_model,
                                X,
                                y_encoded,
                                cv=skf,
                                scoring=scoring)
        # print(f"{metric_name.capitalize()}: {scores.mean():.4f}")
    # print the results for accuracy, recall, precision, and F1 score
    for metric in scoring.keys():
        scores = cv_results[f'test_{metric}']
        print(f"{metric.capitalize()}: {scores.mean():.4f}")
```

```python
# Let's look at the feature importance
# ----------

def plot_top_features(model_name, importances, X, xlim):
    top_features = pd.DataFrame({'Feature': X.columns,
                                 'Importance': importances
                                }).nlargest(10, 'Importance')
    # plot
    sns.set_context("notebook", font_scale=.8)
    plt.figure(figsize=(4, 2.5))
    plt.barh(top_features['Feature'], top_features['Importance'], align='center')
    plt.xlabel('Importance')
    plt.ylabel('Features')
    plt.title(f'Top 10 Features ({model_name})')
    plt.gca().invert_yaxis()
    plt.xlim(0, xlim)
    plt.show()
```

```python
# Logistic Regression
lr_model = model_results['Logistic Regression']['model']
lr_coefficients = np.abs(lr_model.coef_[0])
lr_importance = lr_coefficients / lr_coefficients.max()  # normalize
plot_top_features("Logistic Regression", lr_importance, X, xlim=1)

lr_top_features = pd.DataFrame({'Feature': X.columns,
                                'Importance': lr_importance
                               }).nlargest(10, 'Importance')

# Random Forest
rf_model = model_results['Random Forest']['model']
rf_importance = rf_model.feature_importances_
rf_importance = rf_importance / rf_importance.max()  # normalize
plot_top_features("Random Forest", rf_importance, X, xlim=1)

rf_top_features = pd.DataFrame({'Feature': X.columns,
                                'Importance': rf_importance
                                }).nlargest(10, 'Importance')

# XGBoost
xgb_model = model_results['XGBoost']['model']
xgb_importance = xgb_model.feature_importances_
xgb_importance = xgb_importance / xgb_importance.max()  # normalize
plot_top_features("XGBoost", xgb_importance,X, xlim=1)

xgb_top_features = pd.DataFrame({'Feature': X.columns,
                                'Importance': xgb_importance
                                }).nlargest(10, 'Importance')

# combine all top 10 features
all_features = pd.concat([lr_top_features,
                          rf_top_features,
                          xgb_top_features])['Feature'].unique()
comparison_df = pd.DataFrame({'Feature': all_features})

# add importance values for each model
comparison_df['Logistic Regression'] = comparison_df['Feature'].apply(
    lambda x: lr_importance[X.columns.get_loc(x)] if x in X.columns else 0)

comparison_df['Random Forest'] = comparison_df['Feature'].apply(
    lambda x: rf_importance[X.columns.get_loc(x)] if x in X.columns else 0)

comparison_df['XGBoost'] = comparison_df['Feature'].apply(
    lambda x: xgb_importance[X.columns.get_loc(x)] if x in X.columns else 0)

# plotting
comparison_df.set_index('Feature', inplace=True)
comparison_df.plot(kind='barh', figsize=(12, 8))

# add labels and title
plt.xlabel('Normalized Importance')
plt.ylabel('Features')
plt.title('Feature Importance Comparison Across Models')
plt.gca().invert_yaxis()
plt.legend(title='Model')
plt.show()
```

## 5. Discussion Questions

**1. Why do we need metrics beyond accuracy when using machine learning in the social sciences and public policy?**

Different metrics emphasize various aspects of model performance, so it's essential to choose the one that align with the specific purpose of the analysis. In this project, for instance, the primary purpose is to identify potential high-risk establishments likely to fail audits. This prioritizes minimizing false negatives-cases where high-risk establishments are incorrectly predicted as low risk-better protect public health.   
Relying on accuracy alone ignores the societal cost of misclassification, such as failing to flag dangerous establishments. Metrics like recall and precision help account for these trade-offs and allow for nuanced evaluation of model fairness across different establishment types, ensuring the analysis remains socially equitable and effective in the long term.

**2. Imagine that establishments learned about the algorithm being used to determine who gets audited and they started adjusting their behavior (and changing certain key features about themselves that were important for the prediction) to avoid detection. How could policymakers address this interplay between algorithmic decisionmaking and real world behavior?**

- Diversify Audit Criteria: Avoid over-reliance on a small and fixed set of features by constantly incorporating broader range of indicators. Random audits will also help introduce flexibility of audit pattern, making it harder to predict and manipulate the system.
- Transparency with Caution: Providing limited transparency about how the audit system works. Sharing high-level criteria can encourage compliance without revealing too much exact details and predictive features that could be exploited.
- Dynamic Model Updates: Regularly retrain the model using updated data to account for changes in behavior.
- Enhance Feedback Mechanisms: Combine algorithmic insights with external signals like whistleblower complaints or unexpected market trends to adapt the model to real-world dynamics while maintaining fairness.
